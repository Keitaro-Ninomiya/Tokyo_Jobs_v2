# Split a book by middle line
- **HoriPart Function**

    Detects a line splitting a double sided book in half.
    
    **Variables**
    
    *path*: Path for image of a book
    
    Output: Single candidate for the vertical line spliting a double sided book.

- **CheckLine Function**
    
    Projects a line on a image.

    **Variables**

    *path*: Path for image of a book
    
    *CenterLine* :A list with coordinates of line. List is formated in sequential order; [X axis of starting point, Y axis of starting point,X axis of ending point, Y axis of ending point]
    
    Output: None

```python
CenterLine=HoriPart(path)
CheckLine(path,CenterLine)
```

# Detecting split **within** a page

- **VertPart Function**

    Detects a line splitting a page into two.
    
    **Variables**
    
    *path*: Path for image of a book
    
    Output: Single candidate for the horizontal line spliting a double sided book.

```python
CenterLine=VertPart(path)
CheckLine(path,CenterLine)
```
