# Renaming PNGs

```python
import os
def Rename(origin,path,StrPage):
    filepath=os.path.join(origin,path)
    os.chdir(filepath)
    for Page,file in zip(range(StrPage,StrPage+len(os.listdir(filepath))),os.listdir(filepath)):
        print(Page,file)
        os.rename(file,'Page'+"{:03d}".format(Page)+'.jpg')
```
