```python
import os
import pandas as pd
origin='C:/Users/Keitaro Ninomiya/Box/Research Notes (keitaro2@illinois.edu)/'
Year=1950

filepath=os.path.join(origin,'Tokyo_Jobs/Processed_Data',str(Year),'BPO')
```

```python
def GetOfficePath(origin,Year):
    filepath=os.path.join(origin,'Tokyo_Jobs/Processed_Data',str(Year),'BPO')
    PathList=[]
    for Dept in os.listdir(filepath):
        for Office in os.listdir(os.path.join(filepath,Dept)):
            Test=os.listdir(os.path.join(filepath,Dept,Office))[0]
            TestPath=os.path.join(filepath,Dept,Office,Test)
            if os.path.isdir(TestPath):
                for SubOffice in os.listdir(os.path.join(filepath,Dept,Office)):                                
                    Test=os.listdir(os.path.join(filepath,Dept,Office,SubOffice))[0]
                    TestPath=os.path.join(filepath,Dept,Office,SubOffice,Test)
                    if os.path.isdir(TestPath):
                        for SubSubOffice in os.listdir(os.path.join(filepath,Dept,Office,SubOffice)):                    
                            Finpath=os.path.join(filepath,Dept,Office,SubOffice,SubSubOffice)
                            PathList.append({'Path':Finpath,'Elements':[{'Dept':Dept,'Office':Office,'SubOffice':SubOffice,'SubSubOffice':SubSubOffice}]})
                    else:
                        Finpath=os.path.join(filepath,Dept,Office,SubOffice)
                        PathList.append({'Path':Finpath,'Elements':[{'Dept':Dept,'Office':Office,'SubOffice':SubOffice,'SubSubOffice':None}]})
            else:
                Finpath=os.path.join(filepath,Dept,Office)
                PathList.append({'Path':Finpath,'Elements':[{'Dept':Dept,'Office':Office,'SubOffice':None,'SubSubOffice':None}]})
    return(PathList)
```

```python
def GetPosi(FileName):
    if '主事' in FileName:
        return('主事')
    elif '技師1' in FileName:
        return('技師1')
    elif '技師2' in FileName:
        return('技師2')
    else:
        return('Unknown')
```

```python
DF=pd.DataFrame(columns = ["Year", "Wage","Posi", "Name","Dept","Office","SubOffice","SubSubOffice"])
ErrorList={}
ErrorList[str(Year)]=[]
for n in range(1,len(GetOfficePath(origin,Year))):
    
    try:    
        OfficePath=GetOfficePath(origin,Year)[n]['Path']
        Elements=GetOfficePath(origin,Year)[n]['Elements'][0]
        os.chdir(OfficePath)
        for file in os.listdir(OfficePath):
            Posi=GetPosi(file)
            try:        
                data=pd.read_csv(os.path.join(OfficePath,file), encoding="shift-jis")
            except:
                data=pd.read_csv(os.path.join(OfficePath,file), encoding="shift_jisx0213")
            data=data.rename(columns={'掛': "Unit", '課': "Office", '局': "Dept", 'ページ':'Page', '年度':'Year','賃金':'Wage','名前':'Name'})
            data=data[['Year','Wage','Name']]
            data['Posi'],data['Dept'],data['Office'],data['SubOffice'],data['SubSubOffice']=Posi,Elements['Dept'],Elements['Office'],Elements['SubOffice'],Elements['SubSubOffice']
            DF=pd.concat([DF,data])
    except:
        print('Error at'+OfficePath)
        ErrorList[str(Year)].append(OfficePath)
        continue
```

```python
DF.to_csv(os.path.join(origin,'Tokyo_Jobs/Processed_Data',str(Year),'BPO/Data.csv'), index = False)
```
