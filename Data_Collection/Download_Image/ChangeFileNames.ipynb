```python
import os
import re
User = os.environ.get('USERNAME')  # For Windows
Year = 1940

# Define the path to the directory containing the files
folder_path = f'C:/Users/{User}/Box/Research Notes (keitaro2@illinois.edu)/Tokyo_Jobs/Raw_Data/Metropolitan_DA/{Year}/HQ'

# Change the working directory to the folder path
os.chdir(folder_path)

# List all files in the directory
files = os.listdir()

# Loop over each file in the directory
for file in files:
    # Use a regular expression to find the page number in the file name
    match = re.search(r'_(\d+)$', file)
    if match:
    # Extract the page number
        page_number = match.group(1)
        # Define the new file name using just the page number
        new_name = f"{page_number}.pdf"  # Change the extension if needed
        # Rename the file
        os.rename(file, new_name)
        print(f"Renamed {file} to {new_name}")
```
